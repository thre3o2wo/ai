ê¸°ë³¸ì ìœ¼ë¡œ OpenAI APIëŠ” ìš”ì²­ì— ëŒ€í•œ ì™„ë£Œëœ ë‹µë³€ì„ í•œêº¼ë²ˆì— ë°˜í™˜í•©ë‹ˆë‹¤. ê·¸ëŸ¬ë‚˜ ê¸´ ë‹µë³€ì˜ ê²½ìš° ìŠ¤íŠ¸ë¦¬ë°ì„ ì‚¬ìš©í•˜ë©´ ë§ˆì¹˜ íƒ€ì´í•‘ì„ í•˜ë“¯ì´ í† í° ë‹¨ìœ„ë¡œ ì°¨ë¡€ë¡œ ì‘ë‹µì„ ë°›ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ìŠ¤íŠ¸ë¦¬ë°ì„ í™œìš©í•˜ë©´ ì‚¬ìš©ìì—ê²Œ ì‹¤ì‹œê°„ìœ¼ë¡œ ì‘ë‹µì„ í‘œì‹œí•˜ê±°ë‚˜, ë§¤ìš° ê¸´ ì‘ë‹µì„ ë¶€ë¶„ ë¶€ë¶„ ì²˜ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### ìŠ¤íŠ¸ë¦¬ë°ì´ í•„ìš”í•œ ê²½ìš°
- ì‹¤ì‹œê°„ í”¼ë“œë°±: ì‚¬ìš©ì ê²½í—˜ì„ ê°œì„ í•˜ê¸° ìœ„í•´ ë‹µë³€ ìƒì„±ì„ ê¸°ë‹¤ë¦¬ëŠ” ë™ì•ˆ ì‹¤ì‹œê°„ìœ¼ë¡œ í…ìŠ¤íŠ¸ë¥¼ ë³´ì—¬ì¤„ ë•Œ.
- ê¸´ ì‘ë‹µ ì²˜ë¦¬: ì‘ë‹µì´ ê¸¸ì–´ì„œ í•œêº¼ë²ˆì— ë°›ìœ¼ë©´ ë©”ëª¨ë¦¬ ì‚¬ìš©ì´ ë§ì„ ë•Œ, í† í°ì´ ë„ì°©í•˜ëŠ” ëŒ€ë¡œ ì²˜ë¦¬ ê°€ëŠ¥.
- ì¤‘ê°„ ì‘ì—… ê°€ëŠ¥: ì‘ë‹µì„ ë°›ëŠ” ë„ì¤‘ì—ë„ ë‹¤ë¥¸ ì´ë²¤íŠ¸ë¥¼ ì²˜ë¦¬í•˜ê±°ë‚˜ UI ì—…ë°ì´íŠ¸ë¥¼ í•  ìˆ˜ ìˆìŒ.

### ìŠ¤íŠ¸ë¦¬ë° ì‚¬ìš© ë°©ë²•
OpenAI íŒŒì´ì¬ ë¼ì´ë¸ŒëŸ¬ë¦¬ì—ì„œ ìŠ¤íŠ¸ë¦¬ë°ì„ ì‚¬ìš©í•˜ë ¤ë©´ ìš”ì²­ ì‹œ stream=True ì˜µì…˜ì„ ì£¼ë©´ ë©ë‹ˆë‹¤. ê·¸ëŸ¬ë©´ ì‘ë‹µ ê°ì²´ ëŒ€ì‹  **ì´í„°ë ˆì´í„°(iterator)**ë¥¼ ë°˜í™˜í•˜ë©°, ì´ ì´í„°ë ˆì´í„°ë¥¼ ìˆœíšŒ(for ë¬¸ ë“±)í•˜ë©´ì„œ ë¶€ë¶„ ì‘ë‹µ(chunk)ì„ ë°›ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

OpenAI Chat Completions APIë¥¼ ì‚¬ìš©í•˜ë©´ ë¹„êµì  ì ì€ ì½”ë“œë¡œ ëŒ€í™”í˜• ì±—ë´‡ì„ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” ì½˜ì†”ì—ì„œ ì‚¬ìš©ìì˜ ì…ë ¥ì„ ë°›ì•„ ëª¨ë¸ì˜ ì‘ë‹µì„ ì¶œë ¥í•˜ëŠ” ê°„ë‹¨í•œ ì±—ë´‡ì„ êµ¬í˜„í•´ë´…ë‹ˆë‹¤. ì´ ì±—ë´‡ì€ ì´ì „ ëŒ€í™” ë§¥ë½ì„ ê¸°ì–µí•˜ì—¬ ì—°ì†ì ì¸ ëŒ€í™”ë¥¼ ì£¼ê³ ë°›ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.


````python
cnt = 0
messages=[
    {'role':'system', 'content':'ë„ˆëŠ” ë‚´ ì¶©ì‹¤í•˜ê³  ì¹œì ˆí•œ ë¹„ì„œì•¼.'},
    {'role':'user', 'content':'2020ë…„ ì›”ë“œì‹œë¦¬ì¦ˆì—ì„œ ëˆ„ê°€ ìš°ìŠ¹í–ˆì§€?'},
    {'role':'assistant', 'content':'ë¡œìŠ¤ì•¤ì ¤ë ˆìŠ¤ ë‹¤ì €ìŠ¤ê°€ ìš°ìŠ¹í–ˆìŠµë‹ˆë‹¤. ë‹¤ì €ìŠ¤ëŠ” íƒ¬íŒŒë² ì´ ë ˆì´ìŠ¤ë¥¼ êº¾ê³  ìš°ìŠ¹ì„ ì°¨ì§€í–ˆìŠµë‹ˆë‹¤.'},
]
while True: 
    try:
        prompt = input("ë§ì”€í•˜ì„¸ìš”. (ì¢…ë£Œ í¬ë§ ì‹œ 'exit' ì…ë ¥) : ")
        if prompt.lower()=='exit':
            raise Exception('ëŒ€í™” ì¢…ë£Œ')
        messages.append({'role':'user', 'content':prompt})
        response = client.chat.completions.create(
            model='gpt-4.1-nano',
            messages=messages,
            temperature=1,
            frequency_penalty=0.5
        )
        cnt += 1
        result=response.choices[0].message.content
        print(f'({cnt}), {result}')
        messages.append({'role':'assistant', 'content':result})
    except OpenAIError as e:
        print('ìš”ì²­ëŸ‰ ì œí•œ ì´ˆê³¼ê±°ë‚˜, ì‹œê°„ ì´ˆê³¼, ë„¤íŠ¸ì›Œí¬ ì˜¤ë¥˜. ì±„íŒ… ì…€ì„ ì¬ì‹œì‘í•´ ì£¼ì„¸ìš”.')
        print(e)
        break
````
