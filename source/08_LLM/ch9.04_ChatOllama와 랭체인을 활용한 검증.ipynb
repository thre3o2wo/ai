# 2. LLM 답변 생성
- ollama.com에서 모델 선택 (llama/exaone)
    - powershell 이나 cmd에서 ollama pull llama3.2:1b / exaone3.5:2.4b

```python
from langchain_ollama import ChatOllama
from langchain_core.messages import SystemMessage, HumanMessage

llm = ChatOllama(
    model="exaone3.5:2.4b"
)
messages = [
    SystemMessage(content="너는 친절히 대답해 주는 비서야"),
    HumanMessage(content="2020년 월드시리즈 우승팀은?")
]
ai_message = llm.invoke(input=messages)
print(ai_message.content)
```
