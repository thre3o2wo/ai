```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser
#[cmd] ollama pull exaone3.5:2.4b
llm = ChatOllama(model="exaone3.5:2.4b")
```

```python
# Food Chain
food_chatPrompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a sincere advisor with expertise in world culture and cuisine."),
    ('human', "What is the most popular food in Italy?"),
    ("ai", "Pasta."),
    ('human', "What is the most popular food in Japan?"),
    ('ai', "Sushi."),
    ('human', "What is the most popular food in Mexico?"),
    ('ai', "Tacos."),
    ('human', "What is the most popular food in Hungary?"),
    ('ai', "Goulash."),
    ('human', "What is the most popular food in {country}? Return only the name of a single food.")
])
country = input('어느 나라 음식이 궁금해? ')
output_parser = StrOutputParser()
food_chain = food_chatPrompt_template | llm | output_parser
food = food_chain.invoke({'country':country})
print(food)
```

```python
recipe_chatPrompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a sincere advisor with expertise in world culture and cuisine."),
    ('human', "간단한 파스타 레시피를 알려 주세요."),
    ("ai", """물 끓이기: 큰 냄비에 물을 넉넉히 넣고 센 불로 완전히 끓입니다. 소금을 충분히 넣어 간을 맞춥니다.
파스타 넣기: 마른 파스타를 끓는 물에 넣고, 초반에만 가끔 저어 붙지 않게 합니다.
익히기: 포장지의 조리 시간을 참고하여, 알 덴테(심이 살아 있게) 상태로 익힙니다.
소스 준비 (선택): 파스타가 익는 동안, 별도 팬에 올리브 오일과 다진 마늘을 볶아 소스를 준비합니다.
물 남기기: 파스타 물을 버리기 전에 한 컵 정도 따로 둡니다.
섞기: 파스타의 물기를 빼고(조금 남겨서), 즉시 소스 팬에 옮겨 담습니다.
마무리: 소스와 파스타를 잘 섞고, 농도가 부족하면 남겨둔 파스타 물을 조금씩 넣습니다. 취향에 따라 치즈나 허브를 뿌려 완성합니다."""),
    ('human', "간단한 {food} 레시피를 알려 주세요.")
])
recipe_chain = recipe_chatPrompt_template | llm | output_parser
# food_temp = input('어느 음식 레시피가 궁금해? ')
recipe = recipe_chain.invoke({'food':food})
print(recipe)
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
food_recipe_chain = {'country': RunnablePassthrough()} | {'food':food_chain} | recipe_chain
```

```python
final_prompt = input('어느 나라의 대표 음식 조리법을 가르쳐 줄까요? ')
final_result = food_recipe_chain.invoke(final_prompt)
print(final_result)
```
