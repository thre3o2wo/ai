# 9. 데이터 그룹화 및 집계

```python
from statsmodels.api import datasets
iris = datasets.get_rdataset('iris').data
print(iris.shape)
iris.iloc[[0,49,50,99,100,149]]
```

## 9.1. groupby
cf. pivot_table

```python
#1. 그룹화 방법 1
#    df.groupby(by=[그루핑할 열], as_index=False)[함수 적용해 출력할 열들].기초통계함수()
#    [[함수 적용해 출력할 열들]] 생략 시 모든 열에 적용됨
iris_grouped = iris.groupby(by='Species')
iris_grouped = iris.groupby(by=iris.Species)
#list(iris_grouped) setosa 별 데이터, versicolor 별 데이터 ...
iris_grouped_sum = iris_grouped.sum()
iris_grouped_sum
#iris.groupby(by='Species')[['Sepal.Length','Sepal.Width']].sum()
#iris.groupby(by='Species').sum()
```

```python
# setosa 종과 virginica 종의 종별 합계
iris_grouped_sum.loc[['setosa', 'virginica']]
```

```python
iris_grouped_sum.reset_index(inplace=True) #새로운 인덱스 리셋
iris_grouped_sum
```

```python
# setosa 종과 virginica 종의 종별 합계
iris_grouped_sum[iris_grouped_sum['Species'].isin(['setosa', 'virginica'])]
```

```python
iris_grouped_sum[(iris_grouped_sum['Species']=='setosa' ) | 
                 (iris_grouped_sum['Species']=='virginica' )]
```

```python
#1-1. 그룹화 방법 1-1 : 그루핑할 때 index를 컬럼으로 편입
#    df.groupby(by=[그루핑할 열], as_index=False)[함수 적용해 출력할 열들].기초통계함수()
iris.groupby('Species', as_index=False).sum()
```

```python
#2. 그룹화 방법 2: 피벗 테이블
#    df.pivot_table(values=[함수 적용해 출력할 열들], index='그루핑할 열', aggfunc='함수')
iris.pivot_table(index='Species', 
                 #values=['Sepal.Length', 'Petal.Length'], 
                 aggfunc='mean')
```

### 9.1.1. 다중열 그룹화

```python
#다중열로 그룹화하기 위해 종별 25개씩 나눠 구분하기 위한 열 데이터 추가(num 필드)
#0-25, 1-25, 2-25, 3-25, 4-25, 5-25
import numpy as np
iris['num'] = np.array([[i]*25 for i in range(6)]).ravel()
iris.iloc[::10]
```

```python
# Species별, num별 합산
r = iris.groupby(by=['Species', 'num']).sum()
r
```

```python
r.loc[['setosa','virginica']]
```

```python
r.loc[[('setosa', 0), ('virginica', 4)]]
```

```python
# Species별, num별 합산 (그루핑할 열을 컬럼 데이터로)
r = iris.groupby(by=['Species', 'num'], as_index=False)[['Sepal.Length', 'Sepal.Width']].sum()
r
```

```python
iris.pivot_table(index=['Species', 'num'],
                values=['Sepal.Length', 'Sepal.Width'],
                aggfunc='sum')
```

```python
iris.groupby('Species')[['Sepal.Length',
                         'Sepal.Width',
                         'Petal.Length',
                         'Petal.Width']].describe().T
```

```python
g_sum = iris.groupby('Species')[['Sepal.Length',
                                 'Sepal.Width',
                                 'Petal.Length',
                                 'Petal.Width']].sum().T
g_sum
```

# 10. 데이터 구조 변경
- 와이드 포맷 ↔ 롱 포맷
- Pandas 공식 문서 melt() and wide_to_long() : https://pandas.pydata.org/docs/user_guide/reshaping.html#melt-and-wide-to-long
- Tidy Data : https://vita.had.co.nz/papers/tidy-data.pdf

```python
from statsmodels.api import datasets
airquality = datasets.get_rdataset('airquality').data
airquality.head()
```

## 10.1. melt()를 이용한 언피버팅(Wide → Long)

```python
#melt 하지 않을, 기준이 될 요소 identifier variables를 id_vars에 할당
#airquality.melt(id_vars=['Month', 'Day'])
#airquality.melt(id_vars=['Month', 'Day']).sort_values(by=['Month','Day'])
airquality_melted = airquality.melt(id_vars=['Month', 'Day'],
                                    var_name = '변수명',
                                    value_name = '값').sort_values(by=['Month','Day']).reset_index()
#airquality_melted.reset_index(drop=True, inplace=True)
airquality_melted
```

## 10.2. pivot_table을 이용한 피버팅(Long → Wide)

```python
airquality2 = airquality_melted.pivot_table(index=['Month','Day'],
                                            columns='변수명',
                                            values='값')
airquality2.reset_index(inplace=True)
airquality2.head()
```

```python
airquality2.columns.name
```

```python
airquality2.columns.name = None
airquality2
```

# 11. 데이터프레임에 함수 적용시키기
- apply(func, axis) ** : 데이터프레임이나 시리즈 모두 적용 가능 (1차원이면 axis x)
- applymap(func) : 각 요소별로 적용(데이터프레임에서만 가능)
- map(func) ** : 각 요소별로 적용(1차원 시리즈에서만 가능)

## 11.1. apply() / map()

```python
#Ex.1 1차원 시리즈에 apply 사용
import pandas as pd
member = pd.read_csv('data/ch11_member.csv', comment='#')
member
```

```python
#1999-01-30 문자를 1999 숫자로 return
# def parse_year(datestr):
#     return int(datestr.split('-')[0])
parse_year = lambda datestr : int(datestr.split('-')[0])
parse_year("1999-01-30")
```

```python
#1999-01-30 문자를 1 숫자로 return
# def parse_month(datestr):
#     return int(datestr.split('-')[1])
parse_month = lambda datestr : int(datestr.split('-')[1])
parse_month("1999-01-30")
```

```python
#1999-01-30 문자를 30 숫자로 return
# def parse_day(datestr):
#     return int(datestr.split('-')[2])
parse_day = lambda datestr : int(datestr.split('-')[2])
parse_day("1999-01-30")
```

```python
member['연'] = member['Birth'].apply(lambda datestr : int(datestr.split('-')[0]))
member['월'] = member['Birth'].apply(lambda datestr : int(datestr.split('-')[1]))
member['일'] = member['Birth'].map(lambda datestr : int(datestr.split('-')[2])) #Birth 시리즈만 쓰니 map도 가능해!
member
```

```python
member.info()
```

# ※ melt() 및 함수 적용하기 연습예제 (Wide → Long, pd.concat)
- ✅ 와이드 → 롱 포맷
- 2013년12월 → 2013과 12 나눠 컬럼 구분 (apply, map)
- 컬럼명 조정(df.columns)
- 평당 분양가 → 제곱미터당
- df_last 규모 구분 모든면적만 뺌
- pd.concat

```python
# 2015년 10월~2024년 부동산 데이터
df_last = pd.read_csv(r'C:\ai\downloads\부동산\주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20241231.csv', 
                      encoding='cp949')
df_last.head()
```

```python
# 2013년 9월 ~ 2015년 8월 부동산 데이터
df_first = pd.read_csv(r'C:\ai\downloads\부동산\전국 평균 평당 분양가격(2013년 9월부터 2015년 8월까지).csv', 
                      encoding='cp949')
df_first.head()
```

```python
df_first_melted
```

```python
'2024년10월'.find('월')
```

```python
df_first_melted = df_first.melt(id_vars='지역')
df_first_melted.columns = ['지역명', '연월', '평당분양가']
df_first_melted['연도'] = df_first_melted.apply(lambda datestr : int(datestr['연월'][0:4]), axis=1)
df_first_melted['월'] = df_first_melted.apply(lambda datestr : int(datestr['연월'][5:-1]), axis=1)
df_first_melted['분양가격(제곱미터)'] = df_first_melted['평당분양가'].apply(lambda df : df/3.3)
df_first_melted
```

```python
new_first = df_first_melted.drop(['연월','평당분양가'], axis=1)
new_first.head()
```

```python
new_last = df_last.loc[df_last['규모구분'] == '모든면적'].drop(['규모구분'], axis=1)
new_last
```

```python
df_whole = pd.concat([new_first, new_last]).sort_values(['지역명','연도','월']).reset_index(drop=True)
df_whole
```
