```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
driver.get("http://www.python.org")
```

```python
# 검색창
elem = driver.find_element(By.NAME, 'q') # == select(input([name='q']))
# 비우기
elem.clear() 
# 입력
elem.send_keys('pycon')
```

```python
from selenium.webdriver.common.keys import Keys
# Enter 입력
elem.send_keys(Keys.RETURN) 
```

```python
btn_elem = driver.find_element(By.CSS_SELECTOR, 'button#submit')
# 검색 버튼 클릭
btn_elem.click()
```

```python
result_list = driver.find_elements(By.CSS_SELECTOR, 'li > h3 > a')
for result in result_list:
    title = result.text
    link = result.get_attribute('href') # 정적 방식의 .attrs('href')랑 비슷하지?
    print('{} - {}'.format(title, link))
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
result_list = soup.select('li > h3 > a')
for result in result_list: #정적 방식, url이 제대로 안 먹히네
    print("{} - {}".format(result.text, result.attrs['href']))
```

```python
from urllib.parse import urlparse
current_url = driver.current_url #셀레니엄 통해 접근한 url
print(current_url)
parse_url = urlparse(current_url)
print(parse_url)
domain = f'{parse_url.scheme}://{parse_url.netloc}'
```

```python
result_list = soup.select('li > h3 > a')
for result in result_list: 
    print("{} - {}".format(result.text, domain+result.attrs['href']))
```

```python
driver.close() #브라우저 종료
```

## 2.1. 다음 뉴스 검색

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
#페이지 띄우고
driver = webdriver.Chrome()
driver.get('https://www.daum.net/')
#0.5초 멈춤
time.sleep(0.5)
#검색창 클릭
driver.find_element(By.CLASS_NAME, 'tf_keyword').click
#쿼리 입력
query = input('검색할 단어는?')
driver.find_element(By.CSS_SELECTOR, 'input[type="text"]').send_keys(query)
#검색 버튼 클릭
driver.find_element(By.CLASS_NAME, 'btn_ksearch').click()
```

```python
#뉴스 탭 클릭
# driver.find_elements(By.CSS_SELECTOR, 'ul.list_tab > li')[1].click()
driver.find_element(By.LINK_TEXT,'뉴스').click()
```

```python
#뉴스 제목과 link를 list로 추가
news_list = []
strongs = driver.find_elements(By.CSS_SELECTOR, 'div.item-title > strong')
for strong in strongs:
    a = strong.find_element(By.TAG_NAME, 'a')
    title = a.text
    link = a.get_attribute('href')
    news_list.append([title, link])
```

```python
#다음 페이지로 이동
page_nav = driver.find_element(By.CSS_SELECTOR, 'div.inner_paging')
next_page = page_nav.find_element(By.LINK_TEXT, "2")
next_page.click()
```

## 2.2. 페이징 처리
- 다음 뉴스 페이징 처리 : 원하는 keyword를 원하는 페이지만큼 추출

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

#페이지 띄우고
driver = webdriver.Chrome()
driver.get('https://www.daum.net/')
#0.5초 멈춤
time.sleep(1)

#검색창 클릭
driver.find_element(By.CLASS_NAME, 'tf_keyword').click
#쿼리 입력
query = '아이돌' #input('검색할 단어는?')
driver.find_element(By.CSS_SELECTOR, 'input[type="text"]').send_keys(query)
#검색 버튼 클릭
driver.find_element(By.CLASS_NAME, 'btn_ksearch').click()
#0.5초 멈춤
time.sleep(1)

#뉴스 탭 이동
driver.find_element(By.LINK_TEXT,'뉴스').click()
#0.5초 멈춤
time.sleep(1)

#뉴스 제목과 link를 list로 추가
news_list = []
pages = 3 #int(input('크롤링 대상 페이지수'))
for page in range(1, pages+1):
    strongs = driver.find_elements(By.CSS_SELECTOR, 'div.item-title > strong')
    for strong in strongs:
        a = strong.find_element(By.TAG_NAME, 'a')
        title = a.text
        link = a.get_attribute('href')
        news_list.append([title, link])
    page_nav = driver.find_element(By.CSS_SELECTOR, 'div.inner_paging')
    next_page = page_nav.find_element(By.LINK_TEXT, str(page+1))
    next_page.click()
    time.sleep(2)
driver.close()

#DF
pd.DataFrame(news_list, columns=['제목','링크']).sample(5)
```

## 2.3. 맞춤법 검사기
- 네이버 맞춤법 검사기 이용

```python
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
driver.get('https://www.naver.com/')
time.sleep(1)
elem = driver.find_element(By.ID, 'query')
elem.send_keys('맞춤법 검사기')
elem.send_keys(Keys.RETURN) #form 태그의 submit 클릭 효과
time.sleep(1)
textarea = driver.find_element(By.CLASS_NAME, 'txt_gray')
textarea.send_keys('어렷어부터 우리집은 가난 했었고, 남 들 다하는 외식 한번 한저기 없었고. 일터에 낙아신 어머니 집에없으면 언제나 나혼자서 끄려먹었었던 라면.')
btn = driver.find_element(By.CLASS_NAME, 'btn_check')
btn.click()
time.sleep(2)
soup = BeautifulSoup(driver.page_source, 'html.parser')
result = soup.select_one('p._result_text.stand_txt').text
print(result)
driver.close()
```

```python
with open('data/ch14_맞춤법검사전.txt', 'r', encoding='utf-8') as f:
    text = f.read()
ready_text = [] # 300자씩 나눠진 text
while(len(text)>200):
    temp = text[:200]
    last_dot_index = temp.rfind('\n')
    ready_text.append(text[:last_dot_index])
    text = text[last_dot_index+1 :]
ready_text.append(text)
print(ready_text)
print([len(ready) for ready in ready_text])
```

```python
driver = webdriver.Chrome()
driver.get('https://www.naver.com/')
time.sleep(1)
elem = driver.find_element(By.ID, 'query')
elem.send_keys('맞춤법 검사기')
elem.send_keys(Keys.RETURN) #form 태그의 submit 클릭 효과
time.sleep(1)

textarea = driver.find_element(By.CLASS_NAME, 'txt_gray')
textarea.send_keys

results = '' #맞춤법 검사 완료된 text
for i, ready in enumerate(ready_text):
    print(f'검사중...{i+1}/{len(ready_text)}')
    textarea.clear()
#    textarea.send_keys(Keys.CONTROL, 'a') #ctrl+a 전체 선택. 하지만 비추
    textarea.send_keys(ready)
    time.sleep(1)
    #클릭
    btn = driver.find_element(By.CLASS_NAME, 'btn_check')
    btn.click()
    time.sleep(2)
    #가져오기
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    result = soup.select_one('p._result_text.stand_txt').text
    if '검사가 불가능합니다' in result:
        results += ready + '\n'
    else:
        results += result + '\n'
    time.sleep(1)
driver.close()
```

```python
# 맞춤법 검사 결과를 파일로 출력
with open('data/ch14_맞춤법검사후.txt', 'w', encoding='utf-8') as f:
    f.write(results)
```

## 2.4. 파파고 번역기

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
```

```python
with open('data/ch14_번역전.txt', 'r', encoding='utf-8') as f:
    text = f.read()
ready_text = [] # 3000자씩 나눠진 text
while(len(text)>3000):
    temp = text[:3000]
    last_dot_index = temp.rfind('\n')
    ready_text.append(text[:last_dot_index])
    text = text[last_dot_index+1 :]
ready_text.append(text)
print(ready_text)
print([len(ready) for ready in ready_text])
```

```python
driver = webdriver.Chrome()
driver.get('https://papago.naver.com/')

textarea = driver.find_element(By.ID, 'txtSource')
textarea.send_keys

results = '' #맞춤법 검사 완료된 text
for i, ready in enumerate(ready_text):
    print(f'검사중...{i+1}/{len(ready_text)}')
    textarea.clear()
    textarea.send_keys(ready)
    time.sleep(2)
    #클릭
    btn = driver.find_element(By.ID, 'btnTranslate')
    btn.click()
    time.sleep(4)
    #가져오기
    result = driver.find_element(By.CSS_SELECTOR, 'div#txtTarget').text
#     soup = BeautifulSoup(driver.page_source, 'html.parser')
#     result = soup.select_one('div#txtTarget').text
    time.sleep(3)
#     print(result)
    results += result + '\n'
    time.sleep(1)
driver.close()
print(results)
```

```python
with open('data/ch14_자동화영어번역본.txt', 'w', encoding='utf-8') as f:
    f.write(results)
```
