# 10. 데이터 구조 변경
- 와이드 포맷 ↔ 롱 포맷
- Pandas 공식 문서 melt() and wide_to_long() : https://pandas.pydata.org/docs/user_guide/reshaping.html#melt-and-wide-to-long
- Tidy Data : https://vita.had.co.nz/papers/tidy-data.pdf

```python
from statsmodels.api import datasets
airquality = datasets.get_rdataset('airquality').data
airquality.head()
```

## 10.1. melt()를 이용한 언피버팅(Wide → Long)

```python
#melt 하지 않을, 기준이 될 요소 identifier variables를 id_vars에 할당
#airquality.melt(id_vars=['Month', 'Day'])
#airquality.melt(id_vars=['Month', 'Day']).sort_values(by=['Month','Day'])
airquality_melted = airquality.melt(id_vars=['Month', 'Day'],
                                    var_name = '변수명',
                                    value_name = '값').sort_values(by=['Month','Day']).reset_index()
#airquality_melted.reset_index(drop=True, inplace=True)
airquality_melted
```

## 10.2. pivot_table을 이용한 피버팅(Long → Wide)

```python
airquality2 = airquality_melted.pivot_table(index=['Month','Day'],
                                            columns='변수명',
                                            values='값')
airquality2.reset_index(inplace=True)
airquality2.head()
```

```python
airquality2.columns.name
```

```python
airquality2.columns.name = None
airquality2
```

```python
#1999-01-30 문자를 1999 숫자로 return
# def parse_year(datestr):
#     return int(datestr.split('-')[0])
parse_year = lambda datestr : int(datestr.split('-')[0])
parse_year("1999-01-30")
```

## 12.1. fillna - 결측치를 어떤 값으로 변경
- 대부분의 결측치 대체법
    - fillna 함수 사용 : 중위값, 평균값, 최빈값, 혹은 이전/다음 값, **1차원만 가능
    - drop(dropna 함수 사용), 그룹별 평균값, 그룹별 중위수 (apply 함수 사용)

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, 3, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                  columns=list('ABCD'))
df
```

```python
# 모든 결측치를 무조건 0으로 대체
df.fillna(0)
```

```python
# 결측치를 이전 행의 값으로 대체
df.fillna(method='ffill')
```

```python
# 결측치를 다음 행의 값으로 대체
df.fillna(method='bfill')
```

```python
# 결측치를 열마다 특정 값으로 대체
values = {'A':99, 'B':88, 'C':77, 'D':66}
df.fillna(value=values)
```

```python
df.mean(axis=0)
```

```python
# 결측치를 열평균으로 대체
df.fillna(value=df.mean(axis=0))
```

```python
# 결측치를 열중위값으로 대체
df.fillna(value=df.median(axis=0), inplace=True)
df
```

### Ex-2. iris 결측치를 종별, 열 평균으로 대체 (apply 적용)

```python
import seaborn as sns
iris = sns.load_dataset('iris')
for col in range(iris.shape[1]-1):
    iris.iloc[random.sample(range(150), 20), col] = np.nan
iris.iloc[::25]
```

```python
iris_mean = iris.groupby('species').mean() #species를 인덱스화
iris_mean
```

```python
iris_mean.loc['setosa','sepal_length']
```

```python
#iris.fillna(value=iris_mean) 2차원 데이터프레임으로 fillna 불가
def fillna_grouped(series):
    '''series의 결측치를 대체한 후, series를 return'''
#    pd.options.mode.copy_on_write = True
#    write 할 때 copy 필요하다는 warning 발생하므로
    series = series.copy() #
    for col, item in series.items():
        if col!='species' and np.isnan(item): #결측치인지 여부
            species = series['species']
            series[col] = iris_mean.loc[species, col] #종별 열 평균값 대체  
    return series
fillna_grouped(iris.iloc[2])
```

```python
iris.apply(fillna_grouped, axis=1)
```

```python
# Gemini 코드
import pandas as pd
import numpy as np
import seaborn as sns
import random

# 1. 데이터 로드 및 결측값 생성
iris = sns.load_dataset('iris')

# 결측값 생성 (20개씩 4개 컬럼)
for col in range(iris.shape[1]-1):
    # random.sample을 사용하여 0부터 149까지 20개의 랜덤 인덱스를 추출
    random_indices = random.sample(range(iris.shape[0]), 20)
    iris.iloc[random_indices, col] = np.nan

# 2. 그룹별 평균 계산 (결측값 대체 기준)
# NaN은 자동으로 제외하고 평균을 계산합니다.
iris_mean = iris.groupby('species').mean(numeric_only=True)


# 3. 그룹별 결측값을 대체하는 함수 정의 (apply(axis=1)에 최적화)
def fillna_grouped_safe(row):
    """
    DataFrame의 행(row)을 받아, 해당 행의 species 그룹 평균으로 결측치를 대체합니다.
    (row: pd.Series 객체)
    """
    # 현재 행의 종(species)을 가져옵니다.
    species = row['species']
    
    # 해당 종의 평균값이 담긴 Series를 mean_values 변수에 할당합니다.
    mean_values = iris_mean.loc[species]
    
    # row를 순회하며 결측치를 대체합니다.
    # 단, 'species' 컬럼은 대체 대상이 아니므로 제외합니다.
    for col in row.index:
        # col이 'species'가 아니고, 값이 결측치(np.nan)인 경우에만 대체합니다.
        # .isna()는 결측치 여부를 확인하는 가장 Pandas다운 방법입니다.
        if col != 'species' and pd.isna(row[col]):
            # 결측치를 해당 종(species)의 평균값으로 대체
            row[col] = mean_values[col]
            
    return row # 수정된 행(row)을 반환


# 4. 결측치 대체 실행
# axis=1을 지정하여 함수가 행 단위로 실행되도록 합니다.
# fillna_grouped_safe 함수가 새로운 DataFrame을 반환합니다.
iris_filled = iris.apply(fillna_grouped_safe, axis=1)


# 5. 결과 확인
print("--- 그룹별 평균 대체된 데이터 (상위 10행) ---")
print(iris_filled.head(10))

print("\n--- 대체 후 결측치 확인 (.sum()이 0이면 성공) ---")
# .isna().sum()은 각 컬럼별 결측치 개수의 합을 나타냅니다.
print(iris_filled.isna().sum())
```
