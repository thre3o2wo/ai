```python
import numpy as np
from tensorflow.keras.models import Sequential # 모델 객체 생성
from tensorflow.keras.layers import Input, Dense # 모델 layer 쌓을 때 필요
import matplotlib.pyplot as plt
```

```python
# 1. dataset
x_data = np.array([[0, 0], 
                   [0, 1], 
                   [1, 0], 
                   [1, 1]])
y_data = np.array([[0], [1], [1], [0]])

# 2. 모델 구성
model = Sequential()
#model.add(Input(shape=2))
model.add(Dense(units=10, #출력 개수
               input_dim=2, #입력 개수 input_shape=(2,)
               activation='relu')) #활성화 함수
model.add(Dense(units=20, activation='relu')) #hidden layer의 input_shape는 넣지 않는다.
model.add(Dense(10, 'relu'))
model.add(Dense(units=1, activation='sigmoid'))
print(model.summary())
```

```python
# 3. 모델 학습 과정 설정
# 손실함수 loss= 'mae'- 회귀분석, 
      # 'binary_crossentropy'- 이진분류(로지스틱회귀분석),
      # 'categorical_crossentropy'- 다중분류(원핫인코딩 한 경우)
      # 'sparse_categorical_crossentropy'- 다중분류(원핫인코딩 안 한 경우)
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['binary_accuracy']) # 2진 분류
```

```python
# 4. 학습
hist = model.fit(x_data, y_data, 
                 #batch_size=2,
                 epochs=300, 
                 verbose=2)
```

```python
# 5. 학습 과정 살펴보기
hist.history.keys()
```

```python
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'], 'b', label='loss')
acc_ax = loss_ax.twinx() # loss_ax와 x축 공유하는 acc_ax
acc_ax.plot(hist.history['binary_accuracy'], 'r', label='binary_accuracy')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('binary_accuracy')
loss_ax.legend(loc='center right')
acc_ax.legend(loc='center left')
plt.show()
```

```python
# 6. 모델 평가하기
score = model.evaluate(x_data, y_data) #loss & metrics(binary_accuracy)
print('loss: ', score[0])
print('accuracy: ', score[1])
```
